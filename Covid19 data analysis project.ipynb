```python
covid.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True,axis=1)
```

```python
covid.head()#printing first five rows
```

```python
covid["Date"]=pd.to_datetime(covid['Date'],format='%Y-%m-%d')
```

```python
covid.head()
```

```python
#active cases
covid['Active_cases']=covid['Confirmed']-(covid['Cured']+covid['Deaths'])
covid.tail()#printing last five rows
```

```python
statewise=pd.pivot_table(covid,values=['Confirmed','Cured','Deaths'],index="State/UnionTerritory",aggfunc=max)
```

```python
#recovery rate
statewise["Recovery_rate"]=statewise['Cured']*100/statewise['Confirmed']
```

```python
#mortality rate
statewise["Mortality_rate"]=statewise['Deaths']*100/statewise['Confirmed']
```

```python
#sorting the confirmed in descending order
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")#cmap- color maps(we can give colors to give good representation)
```

```python
#top 10 states based on the active cases
top10_activecases=covid.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most active cases in India",size=25)
```

```python
ax=sns.barplot(data=top10_activecases.iloc[:10],x='State/UnionTerritory',y='Active_cases',linewidth=2,edgecolor='red')
```

```python
#top 10 states based on the active cases
#this is to avoid the overlapping of states in the previous graph
top10_activecases=covid.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 states with most active cases in India",size=25)
ax=sns.barplot(data=top10_activecases.iloc[:10],x='State/UnionTerritory',y='Active_cases',linewidth=2,edgecolor='red')
plt.xlabel('States')
plt.ylabel('Total Active cases')
plt.show()
```

```python
#top 10 states based on number of deaths
#this is to avoid the overlapping of states in the previous graph
top10_deaths=covid.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(18,5))
plt.title("Top 10 states with most number of deaths in India",size=25)
ax=sns.barplot(data=top10_deaths.iloc[:12],x='State/UnionTerritory',y='Deaths',linewidth=2,edgecolor='black')
plt.xlabel('States')
plt.ylabel('Total Deaths')
plt.show()
```
